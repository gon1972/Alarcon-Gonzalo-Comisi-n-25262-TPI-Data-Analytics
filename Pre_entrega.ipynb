```python
#convierte una columna objeto a flotante
def obj_to_float(dataframe, columna): #funcion que saca el signo $ y la coma ',' del dato y lo convierte a flotante
    if dataframe[columna].dtype == 'object':
      dataframe[columna] = (dataframe[columna].str.replace('$', '')
                          .str.replace(',', '').astype(float))
    return dataframe
```

```python
# -------------------------------------------------
# 3Ô∏è‚É£ Funci√≥n para limpiar texto en columnas tipo string
# -------------------------------------------------
def Normalizar_texto(df):  #funcion que normaliza el texto de las columnas del tipo 'object'
    for col in df.select_dtypes(include="object").columns:
        # Se agrupan las operaciones entre par√©ntesis () para escribirlas en varias l√≠neas
        # Python eval√∫a todo el bloque como una √∫nica expresi√≥n.
        df[col] = (
            df[col]
            .astype(str)                              # Convierte cualquier tipo a string
            # .astype(str)  ‚Üí convierte todo a texto; no tiene par√°metros adicionales.
            .str.strip()                               # Elimina espacios al inicio y final
            # .str.strip() no necesita argumentos; borra espacios en blanco por defecto.
            .str.replace(r"[\u200b\t\r\n]", "", regex=True)
            # .str.replace(patron, reemplazo, regex=True)
            #   patron: expresi√≥n regular que busca caracteres invisibles (\u200b, tabulaciones, saltos)
            #   reemplazo: ""  ‚Üí los elimina
            #   regex=True indica que 'patron' es una expresi√≥n regular.
            .str.replace(" +", " ", regex=True)
            # reemplaza "uno o m√°s espacios consecutivos" por un solo espacio
            .str.title()                               # Convierte a T√≠tulo: "juan p√©rez" ‚Üí "Juan P√©rez"
        )
        return df
```

```python
def Normalizar_fechas(df): #funcion que convierte las columnas que contienen fechas al tipo datetime
  for col in df.columns:
        if "fecha" in col.lower():  # detecta columnas con la palabra "fecha"
            df[col] = pd.to_datetime(df[col], errors="coerce", dayfirst=True)
```

```python
def Normalizar(df):
  Normalizar_texto(df)
  Normalizar_fechas(df)
```

# Etapa 1 **Actividad 2**

Realizar un script b√°sico que calcule las ventas mensuales utilizando
variables y operadores.

```python
ventas_perf = (  #agregamos a ventas_perf la columna ingreso=cantidad*precio
    ventas_clean
    .assign(
        ingreso = ventas_clean["precio"] * ventas_clean["cantidad"]
    )
)
ventas_perf.head(5)
```

```python
totales = {}
for fila in ventas_perf.itertuples(index=False): #recorro por filas ventas_perf
    # Acceso por atributo, con el nombres exacto de las columnas
    # (Si no existe la columna, devuelve NaT/nan)
    fecha = getattr(fila, "fecha_venta", pd.NaT)
    importe = getattr(fila, "ingreso", np.nan)
    #print(fila)
    # Clave (a√±o, mes) y acumulaci√≥n
    clave = (fecha.year, fecha.month)  # p.ej., (2024, 3)
    totales[clave] = totales.get(clave, 0.0) + float(importe)
totales
```

# Actividad 2 Etapa 2

**Transformaci√≥n de Datos**: Aplicar filtros y transformaciones para crear una
tabla de ventas que muestre solo los productos con alto rendimiento.

```python
tabla_por_producto = ( #creo una tabla que agrupa por producto y suma el total de los ingresos , las cantidades y la cantidad de operaciones realizadas
    ventas_perf.groupby("producto", as_index=False)
          .agg(
              ingresos_totales=("ingreso", "sum"),
              unidades_totales=("cantidad", "sum"),
              registros=("producto", "size"),
          )
)
tabla_por_producto.sort_values(by=['ingresos_totales','registros'], ascending=False, inplace=True)
tabla_por_producto.head(10)
```

```python
p80_ingreso = tabla_por_producto["ingresos_totales"].quantile( #me indica el valor de ingresos_totales a partir del cual se concentran el 80% de los ingresos
    q=0.80,                # q indica el percentil deseado (0.80 = 80% de los datos por debajo)
    interpolation="linear" # si el percentil no coincide exactamente con un valor real del dataset,
                           # 'linear' interpola entre los dos valores vecinos.
                           # Ejemplo: si el 80% cae entre 4000 y 5000,
                           # calcula un valor proporcional, por ejemplo 4200.
                           # Otros m√©todos posibles:
                           #  - 'lower': toma el menor de los dos valores (4000)
                           #  - 'higher': toma el mayor (5000)
                           #  - 'nearest': el m√°s cercano al percentil
                           #  - 'midpoint': el punto medio exacto (4500)
)
print(f"Umbral (percentil 80) de ingreso_total: {float(p80_ingreso):,.2f}")
```

```python
ventas_top = ( #creo una tabla que me muestre los productos que concentran el 80% de los ingresos de tabla_por_producto
    tabla_por_producto

    .query(
        "ingresos_totales >= @p80_ingreso",  # expr: filtra filas donde ingreso_total es al menos el umbral del P80
        engine="python"                   # motor de evaluaci√≥n (usar 'numexpr' si lo ten√©s y quer√©s performance)

    )

    .sort_values(
        by=["ingresos_totales", "unidades_totales"],  # primero ordena por ingreso_total, luego desempata por unidades
        ascending=[False, False],          # ambos en orden descendente (mayor ‚Üí menor)
        na_position="last",                # coloca NaN al final (√∫til si alguna m√©trica qued√≥ en NaN)
        ignore_index=True                  # reindexa el resultado secuencialmente (0..n-1)

    )
)
ventas_top
```
